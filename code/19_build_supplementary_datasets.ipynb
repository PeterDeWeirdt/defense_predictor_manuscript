## Table S4. Predictions for 1,000 diverse prokaryotic genomes.
1. Novel seqs and q_cov
2. Novel Pfams
3. Assemblies


```python
refseq1k_predictions = pd.read_parquet('../data3/interim/refseq1k_predicted_defense2.pq')
```

```python
novel_refseq1k_predictions = refseq1k_predictions.query("hit_category == 'Predicted novel defense gene'")
```

```python
novel_refseq1k_pfams = pd.read_csv('../data3/interim/candidate_refseq_pfams.csv')
```

```python
filtered_refseq1k_pfams = (novel_refseq1k_pfams[~novel_refseq1k_pfams['hit_name_Pfam'].str.contains('Cas9')])
```

```python
filtered_refseq1k_pfams
```

```python
refseq1k_assemblies = pd.concat([pd.read_csv('../data/interim/refseq_500_selected_assemblies.csv'), 
                                 pd.read_csv('../data/interim/refseq_1k_selected_assemblies.csv')])
```

```python
refseq1k_assemblies
```

```python
with pd.ExcelWriter('../data3/processed/table S04 - 1k prokaryotes predictions.xlsx') as writer:  
    (novel_refseq1k_predictions
     .drop(columns=[x for x in drop_foldseek_cols if x in novel_refseq1k_predictions.columns])
     .to_excel(writer, sheet_name='novel_clusters', index=False))
    (filtered_refseq1k_pfams
     .drop(columns=[x for x in drop_foldseek_cols if x in filtered_refseq1k_pfams.columns])
     .to_excel(writer, sheet_name='novel_pfams', index=False))
    refseq1k_assemblies.to_excel(writer, sheet_name='assemblies', index=False)
```

## Table S5. Predictions for 69 diverse E. coli strains.
1. Novel proteins and q_cov
2. Pfams
3. Assemblies
4. Known Pfams


```python
novel_ecor_predictions = pd.read_csv('../data3/interim/ecor_unique_novel_proteins.csv')
```

```python
novel_ecor_predictions
```

```python
novel_ecor_pfams = pd.read_csv('../data3/interim/ecor_unique_novel_pfam_domains.csv')
```

```python
novel_ecor_pfams
```

```python
ecor_assemblies = pd.read_csv('../data/raw/vassallo_strains.csv', names=['assembly'])
filtered_ecor_assemblies = ecor_assemblies[~ecor_assemblies['assembly'].isin(['GCF_003333865.1', 'GCF_003334065.1'])]
```

```python
filtered_ecor_assemblies
```

```python
blacklisted_domains = pd.read_csv('../data3/interim/df_blacklisted_domains.csv')
```

```python
with pd.ExcelWriter('../data3/processed/table S05 - 69 E. Coli predictions.xlsx') as writer:  
    (novel_ecor_predictions
     .drop(columns=[x for x in drop_foldseek_cols if x in novel_ecor_predictions.columns] + ['present', 'n_proteins'])
     .to_excel(writer, sheet_name='novel_clusters', index=False))
    (novel_ecor_pfams
     .drop(columns=[x for x in drop_foldseek_cols if x in novel_ecor_pfams.columns])
     .to_excel(writer, sheet_name='pfams', index=False))
    filtered_ecor_assemblies.to_excel(writer, sheet_name='assemblies', index=False)
    blacklisted_domains.to_excel(writer, sheet_name='known_defensive_pfams', index=False)
```
